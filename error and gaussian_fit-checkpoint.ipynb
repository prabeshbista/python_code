```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.special import erf #error function importing from scipy
plt.style.use('seaborn-poster')
from scipy.optimize import curve_fit # curve fit importing from sciy
```

```python
# imorting the data

file_0V = pd.read_csv('0V_chessy_momentum m3.0_400 nm_FA 100_CA open_ext 10k_bias 0V_200 s_1_003 (2).csv' ) #importing a file having OV as biasing
file_1V = pd.read_csv('1V_chessy_momentum_m3.0_400_nm_FA_100_CA_open_ext_10k_bias_1V_200_s_005.csv') #importing a file having 1V as biasing
file_neg_1V = pd.read_csv('-1V_chessy_momentum m3.0_400 nm_FA 100_CA open_ext 10k_bias -1V_200 s_1_009 (2).csv') #importing a file having -1V as biasing
file_2V = pd.read_csv('2V_chessy_momentum m3.0_400 nm_FA 100_CA open_ext 10k_bias 2V_200 s_006.csv') #importing a file having 2V as biasing
file_neg_2V = pd.read_csv('-2V_chessy_momentum m3.0_400 nm_FA 100_CA open_ext 10k_bias -2V_200 s_1_012.csv') #importing a file having -2V as biasing


x= np.linspace(350,378,512 ) # the slice represents the time in ns
```

```python

# =============================================================================
# ploting 0V biasing voltage
# =============================================================================



y_center_0V = (file_0V['Mean_center'])*np.pi*15*15*20  # the time represents the mean count of central part
y_full_0V = (file_0V['Mean_full'])*750*750  # the time represents the mean count of central part


plt.figure(figsize=(10,7))  #  second plot
plt.xlabel('time of flight (ns)')
plt.ylabel('photoemission intensity (counts)')
plt.title('OV ')
plt.plot(x, y_full_0V, 'o', label='full')

plt.plot(x, y_center_0V,'o' ,color='gray', label='center')
plt.ticklabel_format(style='scientific', axis='y', scilimits=(0,0))

# define the funcion to fit the curve for the full circle data set (left side of the curve which denotes the fermi edge)
x_data_0V_full_0V = x[(x>354) & (x<357.7) ]

y_data_0V_full_0V = y_full_0V[(x>354) & (x<357.7)]

def gauss_f_0V(x,A, mu, sig): # gauss_f_OV denotes the gaussuain function having Center at mu with widht of sig. A denotes the amplitude
    return A*np.exp(-(x-mu)**2/sig**2)


#  Fit the error func_0Vtion for full integrated curve
popt_left_0V, pcov_left = curve_fit(gauss_f_0V, x_data_0V_full_0V, y_data_0V_full_0V, p0=(175000,357,4))


# Plot fitted right curve
p3=plt.plot(x_data_0V_full_0V, gauss_f_0V(x_data_0V_full_0V, *popt_left_0V), 'r-')


# Step : Define the func_0Vtion for curve fitting of central curve (right side of curve which defines the vacuum edge)
x_data_0V = x[(x>358.4) & (x<360) ]
y_data_0V = y_center_0V [(x>358.4) & (x<360)]


def func_0V(x, a, b, c, d): 
    return a * erf(b * (x -c)) + d # erf function is used to fit the curve


#  Fit the error func_0Vtion for center curve
popt_right_0V, pcov_right = curve_fit(func_0V, x_data_0V, y_data_0V, p0=(80000,0.05,360,66))



# Plot fitted right curve
p4= plt.plot(x_data_0V, func_0V(x_data_0V, *popt_right_0V), 'r-', label='fit')

#midpoint of vertical line for center curve
V_line = func_0V(x_data_0V, *popt_right_0V)

# Calculate the center y value
center_y_0V = (np.max(V_line) + np.min(V_line)) / 2

# Find the nearest points in y to the center_y_0V
nearest_points = np.abs(V_line - center_y_0V)
index_lower = np.argmin(nearest_points)
index_upper = index_lower + 1 if index_lower < len(V_line) - 1 else index_lower - 1

# Interpolate to find the x coordinate corresponding to center_y_0V
x_lower = x_data_0V[index_lower]
x_upper = x_data_0V[index_upper]
y_lower = V_line[index_lower]
y_upper = V_line[index_upper]

slope = (y_upper - y_lower) / (x_upper - x_lower)
center_x_0V = x_lower + (center_y_0V - y_lower) / slope






plt.legend()
plt.show()
```
