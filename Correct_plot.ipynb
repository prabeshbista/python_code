```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
d1=pd.read_csv('GOLD 1-1000.Dat')
d1.dropna(subset=['WF (mV)'],how='any', inplace=True)

binwidth=5
m_n=np.floor(min(d1['WF (mV)'])/binwidth)*binwidth
m_x=np.floor(max(d1['WF (mV)'])/binwidth)*binwidth+binwidth            
print(m_n)
print(m_x)
print(m_x-m_n)
binnum=int((m_x-m_n)/binwidth)
print(binnum)
a,b,c=plt.hist(d1['WF (mV)'],bins=binnum,range=(m_n,m_x),edgecolor='k')
hf=max(a)/2
plt.axhline(y=hf,color='r')
plt.xticks(b,rotation=45)

print(min(d1['WF (mV)']))
print(max(d1['WF (mV)']))
print(a,b)

# going from point to point
for i in range(0,len(a)-1):
    #to find the left edge of FWHM   
    if a[i]<hf:
        if a[i+1]>=hf:
            leftedge=i
    #to find the  right edge of FWHM
    if a[i]>=hf:
        if a[i+1]<hf:
            rightedge=i
            
#to save the x and y values for the left edge          
leftedge_low_x=b[leftedge]+binwidth/2
leftedge_high_x=b[leftedge+1]+binwidth/2
leftedge_low_y=a[leftedge]
leftedge_high_y=a[leftedge+1]

# line connecting the two points of the left edge.            
plt.plot((leftedge_low_x,leftedge_high_x),(leftedge_low_y,leftedge_high_y))

         
#to save the x and y values for the right edge 
rightedge_low_x=b[rightedge]+binwidth/2
rightedge_high_x=b[rightedge+1]+binwidth/2
rightedge_low_y=a[rightedge]
rightedge_high_y=a[rightedge+1]

#line connecting the two points of the right edge
plt.plot((rightedge_low_x,rightedge_high_x),(rightedge_low_y,rightedge_high_y))

# to find the x value of the left edge 
leftedge_x= ((hf-leftedge_low_y)*(leftedge_high_x-leftedge_low_x))/(leftedge_high_y-leftedge_low_y)+ leftedge_low_x
    
#to plot the vertical line on left edge
plt.axvline(x=leftedge_x,color='r')

# to find the x value of the right edge 
rightedge_x= ((hf-rightedge_low_y)*(rightedge_high_x-rightedge_low_x))/(rightedge_high_y-rightedge_low_y)+ rightedge_low_x
    
#to plot the vertical line on right edge
plt.axvline(x=rightedge_x,color='r')
plt.figure(figsize=(6,6))
    
#to calculate the FWHM
width= rightedge_x-leftedge_x
print(width,'width')

   

```

```python
d1=pd.read_csv('WS2-Pure-1000.Dat')
d1.dropna(subset=['WF (mV)'],how='any', inplace=True)

binwidth=5
m_n=np.floor(min(d1['WF (mV)'])/binwidth)*binwidth
m_x=np.floor(max(d1['WF (mV)'])/binwidth)*binwidth+binwidth            
print(m_n)
print(m_x)
print(m_x-m_n)
binnum=int((m_x-m_n)/binwidth)
print(binnum)
a,b,c=plt.hist(d1['WF (mV)'],bins=binnum,range=(m_n,m_x),edgecolor='k')
hf=max(a)/2
plt.axhline(y=hf,color='r')
plt.xticks(b,rotation=45)
plt.tight_layout()
print(a)


# going from point to point
for i in range(0,len(a)-1):
    #to find the left edge of FWHM   
    if a[i]<hf:
        if a[i+1]>=hf:
            leftedge=i
    #to find the  right edge of FWHM
    if a[i]>=hf:
        if a[i+1]<hf:
            rightedge=i
            
#to save the x and y values for the left edge          
leftedge_low_x=b[leftedge]+binwidth/2
leftedge_high_x=b[leftedge+1]+binwidth/2
leftedge_low_y=a[leftedge]
leftedge_high_y=a[leftedge+1]

# line connecting the two points of the left edge.            
plt.plot((leftedge_low_x,leftedge_high_x),(leftedge_low_y,leftedge_high_y))

         
#to save the x and y values for the right edge 
rightedge_low_x=b[rightedge]+binwidth/2
rightedge_high_x=b[rightedge+1]+binwidth/2
rightedge_low_y=a[rightedge]
rightedge_high_y=a[rightedge+1]

#line connecting the two points of the right edge
plt.plot((rightedge_low_x,rightedge_high_x),(rightedge_low_y,rightedge_high_y))

# to find the x value of the left edge 
leftedge_x= ((hf-leftedge_low_y)*(leftedge_high_x-leftedge_low_x))/(leftedge_high_y-leftedge_low_y)+ leftedge_low_x
    
#to plot the vertical line on left edge
plt.axvline(x=leftedge_x,color='r')

# to find the x value of the right edge 
rightedge_x= ((hf-rightedge_low_y)*(rightedge_high_x-rightedge_low_x))/(rightedge_high_y-rightedge_low_y)+ rightedge_low_x
    
#to plot the vertical line on right edge
plt.axvline(x=rightedge_x,color='r')
    
#to calculate the FWHM
width0= rightedge_x-leftedge_x
print(width0,'width')

   


```
