```python
#wavelength of nopa
x_wavelength = np.array([691, 746,815, 933])  

#power of nopa with wlg in mw
y_power_nopa_pump = np.array([2.94,25.21,15.59,13.92 ])



#error bar of nopa with wlg in mw
y_power_nopa_errorbar = np.array([0.25,0.33,0.44,0.3])

 #power of wlg in mw
y_power_wlg = np.array([0.403,0.418,0.456,0.475])

#error bar of wlg in mw
y_power_wlg_errorbar = np.array([0.006, 0.018, 0.022, 0.008 ]) 

#power of nopa 
y_nopa = y_power_nopa_pump - y_power_wlg

#pulse energy of nopa 
y_pulse_energy_nopa =  y_nopa * 10**(-8)

print("Pulse energy are: ",y_pulse_energy_nopa)

#errorbar of nopa
y_err = y_power_nopa_errorbar + (y_power_wlg_errorbar)

#errorbar of pulse energy
y_err_pulse_energy = y_err * 10**(-8)
 
print(y_err)
print('error bar of pulse energy are:', y_err_pulse_energy)
```

```python
#the graph of power wrt wavelength



plt.errorbar(x_wavelength, y_nopa, fmt='ro', yerr=y_err, elinewidth=1, capsize=7, ecolor='black')


plt.plot(x_wavelength,y_nopa,'ro')
plt.title('power of nopa')
plt.xlabel('''Wavelength (nm)''')
plt.ylabel('Power (mW)')
plt.legend(loc='best')
# plt.show()
```
